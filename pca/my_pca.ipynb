```python
# import  necessary modules
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the digits data
digits = datasets.load_digits()
images, targets = digits.data, digits.target

print(images.shape)      # (1797, 64)
images_ = images.T       # (64, 1797)    here 64 = 8x8
# normalize 
images_mean = np.mean(images_, axis=1)
images_ = images_ - images_mean.reshape(64, 1) 
```

```python
# covariance matrix
cov_var = np.dot(images_, images_.T)
```

```python
# calculate eig_values, eig_vectors
# np.linglg.eig will automatically sort eig_values descently
eig_values, eig_vectors = np.linalg.eig(cov_var)
```

```python
# here we only care the first two eig vectors
# note that eig_vectors is arrange in the second axis
values_selcted, vectors_selected = eig_values[:2], eig_vectors[:, :2]
```

```python
# here are two eig_vectors and each eig_vectors has 64 dims
print(vectors_selected.shape)
```

```python
# get the reduced dimensional coordinates 
res = np.dot(vectors_selected.T, images_)
```

```python
plt.scatter(res[0, :], res[1, :], c=targets)
plt.show()
```
