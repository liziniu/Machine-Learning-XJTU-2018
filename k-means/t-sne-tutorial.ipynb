```python
from sklearn.manifold import TSNE
from sklearn import datasets 
import matplotlib.pyplot as plt
```

```python
# load the digital datasets
digits = datasets.load_digits()   
images, targets = digits.images, digits.target
print(images.shape, targets.shape)
```

```python
plt.imshow(images[2])
plt.show()
print("It is {}".format(targets[2]))
```

```python
# reshape the data into (1797, 64)
import numpy as np
images_reshape = np.transpose(images, [2,1,0])
images_reshape = images_reshape.reshape(64, 1797).T
```

```python
# transfer the high dimensional data(d=64) into low dimensional data(d=2)
tsne = TSNE(n_components=2)
tsne.fit(images_reshape)           #it may take a few minutes
images_embedded = tsne.fit_transform(images_reshape)  
print(images_embedded.shape)
```

```python
# visualize the result
plt.figure()
plt.scatter(images_embedded[:, 0], images_embedded[:, 1], c=targets)
plt.show()
```
